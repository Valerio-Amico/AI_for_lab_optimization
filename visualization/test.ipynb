```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import griddata

# Generate Data
X=np.random.random(10)
Y=np.random.random(10)

Z=X*Y

XY = np.array([X,Y]).T

# Interpolation
# N=100j       
extent=(0,1,0,1) # [min(X), max(X), min(Y), max(Y)]
# xs,ys = np.mgrid[extent[0]:extent[1]:N, extent[2]:extent[3]:N]

N = 10
x, y = np.meshgrid(np.linspace(extent[0],extent[1],N), np.linspace(extent[2],extent[3],N), indexing="xy")
xy = np.array([x.flatten(), y.flatten()]).T

print(XY)
print(Z)
print(xy)
resampled=griddata(XY, Z, xy, method='nearest').reshape([N,N])

#Plot
fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.imshow(resampled, extent=extent, interpolation="lanczos") 

ax.scatter(X,Y,c='r')

plt.show()
```
